# We are trying to detect Rocks and Mines from some SONAR Data

<b>We are going to perform this machine learning task by using the logistic regression model. So, we plan to seperate the given data into some training and testing data. But before that we want to do some necessary preprocessing, find any statistical relations and then train the model based on the training data. Finally we want to test the model using the test data.</b>

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv('sonar data.csv', header = None)
sonar_data.head()
```

```python
# Normally we need to check some statistical measures of the dataset. So we can follow some of the functions below
# to get an idea of what needs to be done.....

#sonar_data.shape() # gives the number of rows and columns
#sonar_data.describe() # gives the statistical values like - percentiles, mean etc 
#sonar_data[60].value_counts() # gives the number of unique item counts in that column specified(in this case rocks and mines)
```

```python
sonar_data.groupby(60).mean() # mean for rocks and mines for each of the features
```

```python
# seperating data and labels
X = sonar_data.drop(columns = 60, axis = 1)
Y = sonar_data[60]
```

```python
# train and test data split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, stratify = Y, random_state = 1)

# test_size = 0.3 means that 30% data is left for testing and rest 70% for training
# stratify = Y means that we are splitting on the basis of the label data stored in Y 
# random_state can be 1,2,... it specifies the pattern of random selection based on the labels
```

```python
# model training
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# model testing
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)

print(test_data_accuracy)
```
